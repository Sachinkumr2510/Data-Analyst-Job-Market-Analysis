```python
import re

def clean_salary(salary):
    if pd.isna(salary) or salary == "-1":  # Handle missing or placeholder values
        return None

    # Remove non-numeric characters except '-' (e.g., "$80K-$90K per year" → "80-90")
    salary = re.sub(r"[^\d\-]", "", salary)

    # Check if we have a valid salary range
    if "-" in salary:
        parts = salary.split("-")
        if len(parts) == 2 and parts[0] and parts[1]:  # Ensure both values exist
            low, high = int(parts[0]) * 1000, int(parts[1]) * 1000
            return (low + high) / 2  # Return average salary

    return None  # Return None if the format is unexpected
```

```python
df["Average Salary"] = df["Salary Estimate"].apply(clean_salary)
```

```python
# 1. Salary Distribution
plt.figure(figsize=(5, 3))
sns.histplot(df["Average Salary"].dropna(), bins=30, kde=True, color="blue")
plt.title("Salary Distribution of Data Analyst Jobs", fontsize=14)
plt.xlabel("Average Salary (USD)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# 2. Top 10 Cities with Most Job Listings
plt.figure(figsize=(5, 3))
top_cities = df["City"].value_counts().head(10)
sns.barplot(x=top_cities.index, y=top_cities.values, hue = top_cities.index, palette = 'viridis', legend = False)
plt.title("Top 10 Cities for Data Analyst Jobs", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Job Listings", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Average Salary by Industry (Top 10)
plt.figure(figsize=(5, 3))
top_industries = df.groupby("Industry")["Average Salary"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_industries.values, y=top_industries.index, hue = top_industries.index, palette="magma")
plt.title("Top 10 Industries Paying the Highest Salaries", fontsize=14)
plt.xlabel("Average Salary (USD)", fontsize=12)
plt.ylabel("Industry", fontsize=12)
plt.show()
```

```python
# 4. Average Salary by State (Top 10)
plt.figure(figsize=(5, 3))
top_states = df.groupby("State")["Average Salary"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_states.values, y=top_states.index, hue = top_states.index, palette="coolwarm")
plt.title("Top 10 States Paying the Highest Salaries", fontsize=14)
plt.xlabel("Average Salary (USD)", fontsize=12)
plt.ylabel("State", fontsize=12)
plt.show()
```

```python
# 5. Relationship Between Company Rating & Salary
plt.figure(figsize=(4, 2))
sns.scatterplot(data=df, x="Rating", y="Average Salary", alpha=0.5)
plt.title("Company Rating vs Average Salary", fontsize=14)
plt.xlabel("Company Rating", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.show()
```
